```python
#122 dispersion curve stuff


import numpy as np
import matplotlib.pyplot as plt
filename = 'dispersionvalues.txt'
import matplotlib

wavelength = (3650, 3663, 4047, 4078, 4358, 4916, 5461, 5770, 5791)
dispersion = (-5, -4, -7, -6, -5, 0, 9, 15, 16)
i=0
err = np.ones(9)
for n in dispersion:
    err[i]=.5
    i+=1
plt.figure(figsize=(15,12))
plt.errorbar(wavelength, dispersion, fmt='b', marker='o', linewidth='0', yerr=err)
vert=np.arange(-10,50,1)

z = np.polyfit(wavelength, dispersion, 2)
print z
x = np.arange(3500, 6000, 1)
def fit(lamb):
    return(z[0] * lamb**2 + z[1] * lamb + z[2])
y = (z[0] * x**2 + z[1] * x + z[2])
plt.plot(x,y, 'g')

chi_2=0
i=0
for n in dispersion:
    theor=fit(wavelength[i])
    chi_2 += (theor-n)**2/.5**2
    i+=1
print chi_2/(i-2)


measurement=4095.347*np.ones(len(vert))

#plt.plot(measurement, vert, 'b-')
disp = z[0]*measurement**2+z[1]*measurement+z[2]

print ''
print 'The dispersion at ', measurement[0], 'is ', disp[0]

print 'So the measurement of ', measurement[0], 'measures a wavelength of ', measurement[0]-disp[0]
plt.grid()
matplotlib.rc('xtick', labelsize=20)
matplotlib.rc('ytick', labelsize=20)
plt.ylabel('Difference from accepted value', fontsize=36)
plt.xlabel('Wavelength in angstroms', fontsize=36)
plt.savefig('dispersion.png')
```
