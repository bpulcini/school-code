```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as op
```

```python
os.getcwd()
```

```python
def f(x, A, b, c):
    return A/(c* np.sqrt(2*np.pi)) * np.exp(-(x - b)**2.0 / (2 * c**2))

def chi_sq(mu_sig, h, x_op):
    chi_2 = ((h - f(x_op, mu_sig[0], mu_sig[1], mu_sig[2]))**2)/0.00001946**2
    return np.sum(chi_2)/(len(x_op)-3)


```

```python
filename1 = 'Alpha2_1s_200mv.txt'
filename2 = 'Beta1_1s_10mv.txt'
filename3 = 'Gamma1_1s_1mv.txt'
filename4 = 'Delta_100microv_1sec.txt'
filename5 = 'Epsilon4_50micov_3s.txt'
filename6 = 'Epsilon3_50microv_1s.txt'
x_a, y_a = np.loadtxt(filename1, usecols=(0,1), unpack=True)

```

```python

plt.figure(figsize=(20,16))


plt.plot(x_a,y_a)




alpha=0.0369# WAVELENGTH DEPENDANT!!!!!

#guesses
A = .1 # WAVELENGTH DEPENDANT!!!!!
mean_op = 6600 # WAVELENGTH DEPENDANT!!!!!
sigma_op = .3 # WAVELENGTH DEPENDANT!!!!!



i=0
left=True
for n in y_a:
    if left ==True:
        if n>alpha:
            left=False
            left_edge = i
    else:
        if n<alpha:
            right_edge = i
            break
    i+=1

x_optimize = x_a[left_edge:right_edge]
y_optimize = y_a[left_edge:right_edge]


x_chi = x_a[0:right_edge]
y_chi = y_a[0:right_edge]


mu_sig = (A, mean_op, sigma_op)
plt.errorbar(x_optimize,y_optimize, yerr=0.00001946, fmt='go')



mu_sig_opt = op.fmin(chi_sq, mu_sig, args=(y_optimize, x_optimize))
print ' '
print 'Chi square Deuterium = ', chi_sq(mu_sig_opt, y_optimize, x_optimize)
print 'Chi square Deuterium = ', chi_sq(mu_sig_opt, y_chi, x_chi)
print ' '
print mu_sig_opt[0]
print mu_sig_opt[1]
print mu_sig_opt[2]

y_fit = f(x_a, mu_sig_opt[0],mu_sig_opt[1],mu_sig_opt[2])

plt.plot(x_a,y_fit, 'r--')


```

```python

plt.figure(figsize=(20,16))


plt.plot(x_a,y_a)
#second peak (Hydrogen)

A2 = .1 # WAVELENGTH DEPENDANT!!!!!
mean_op2 = 6600 # WAVELENGTH DEPENDANT!!!!!
sigma_op2 = .3 # WAVELENGTH DEPENDANT!!!!!
alpha=0.038  # WAVELENGTH DEPENDANT!!!!!

i2=right_edge+2
left2=True
for n in y_a[i2:]:
    if left2 ==True:
        if n>alpha:
            left2=False
            left_edge2 = i2
    else:
        if n<alpha:
            right_edge2 = i2
            break
    i2+=1
    
    
x_optimize2 = x_a[left_edge2:right_edge2]
y_optimize2 = y_a[left_edge2:right_edge2]

plt.errorbar(x_optimize2,y_optimize2, yerr=0.00001946, fmt='ro')

#optimizing

mu_sig2 = (A2, mean_op2, sigma_op2)

print ' '
print ' '

mu_sig_opt2 = op.fmin(chi_sq, mu_sig2, args=(y_optimize2, x_optimize2))




y_fit2 = f(x_a, mu_sig_opt2[0],mu_sig_opt2[1],mu_sig_opt2[2])

plt.plot(x_a,y_fit2, 'g--')


print ' '


print ' '

#new arrays for chi calc
x_chi2 = x_a[left_edge2:-1]
y_chi2 = y_a[left_edge2:-1]

print mu_sig_opt2[0]
print mu_sig_opt2[1]
print mu_sig_opt2[2]
print ' '

print 'Chi square Hydrogen = ', chi_sq(mu_sig_opt2, y_optimize2, x_optimize2)
print 'Chi square Hydrogen = ', chi_sq(mu_sig_opt2, y_chi2, x_chi2)
```
