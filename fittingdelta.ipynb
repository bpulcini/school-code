```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as op

filename1 = 'Delta_100microv_1sec.txt'
x_a, y_a = np.loadtxt(filename1, usecols=(0,1), unpack=True)

plt.figure(figsize=(20,16))
#plt.subplot(121)
plt.plot(x_a,y_a)

#fitting

def f(x, A, b, c):
    return A * np.exp(-(x - b)**2.0 / (2 * c**2))

#plt.plot([4090, 4099], [.000052, .000052])

i=0
left=True
alpha=0.000052
for n in y_a:
    if left ==True:
        if n>alpha:
            print 'Left edge = ', x_a[i], y_a[i]
            left=False
            left_edge = i
    else:
        if n<alpha:
            print 'Right edge = ', x_a[i], y_a[i]
            right_edge = i
            break
    i+=1

x_optimize = x_a[left_edge:right_edge]
y_optimize = y_a[left_edge:right_edge]
#plt.subplot(122)
#plt.plot(x_optimize, y_optimize, 'go')

A = .00003
mean_op = 4094
sigma_op = .1
mu_sig = (A, mean_op, sigma_op)

def chi_sq(mu_sig, h, x_op):
    chi_2 = ((h - f(x_op, mu_sig[0], mu_sig[1], mu_sig[2]))**2)/((2.54*10**-6))
    return np.sum(chi_2)/(len(x_op))

mu_sig_opt = op.fmin(chi_sq, mu_sig, args=(y_optimize, x_optimize))
print mu_sig_opt[0]
print mu_sig_opt[1]
print mu_sig_opt[2]

print chi_sq(mu_sig_opt, y_optimize, x_optimize)


y_fit = f(x_a, mu_sig_opt[0],mu_sig_opt[1],mu_sig_opt[2])
plt.plot(x_a,y_fit, 'r--')

#y_shift=np.zeros(len(y_a))
#for i in xrange(0,len(y_a)):
#    y_shift[i]=y_a[i]-y_fit[i]
#plt.plot(x_a,y_shift,'g--')



#left edge of second peak

i2=right_edge+2
left2=True
alpha = .000052
for n in y_a[i2:]:
    if left2 ==True:
        if n>alpha:
            print 'Left edge = ', x_a[i2], y_a[i2]
            left2=False
            left_edge2 = i2
    else:
        if n<alpha:
            print 'Right edge = ', x_a[i2], y_a[i2]
            right_edge2 = i2
            break
    i2+=1
    
x_optimize2 = x_a[left_edge2:right_edge2]
y_optimize2 = y_a[left_edge2:right_edge2]

#plt.plot(x_optimize2, y_optimize2, 'ro')

#optimizing
A2 = .00003
mean_op2 = 4095
sigma_op2 = .1
mu_sig2 = (A2, mean_op2, sigma_op2)

mu_sig_opt2 = op.fmin(chi_sq, mu_sig2, args=(y_optimize2, x_optimize2))
print mu_sig_opt2[0]
print mu_sig_opt2[1]
print mu_sig_opt2[2]

#print chi_sq(mu_sig_opt2, y_optimize, x_optimize)


y_fit = f(x_a, mu_sig_opt[0],mu_sig_opt[1],mu_sig_opt[2])
plt.plot(x_a,y_fit, 'r--', linewidth=4)

y_fit2 = f(x_a, mu_sig_opt2[0],mu_sig_opt2[1],mu_sig_opt2[2])
plt.plot(x_a,y_fit2, 'r--', linewidth=4)

plt.plot([mu_sig_opt[1], mu_sig_opt[1]],[0,.00008], 'black', linewidth=3)
plt.plot([mu_sig_opt2[1], mu_sig_opt2[1]],[0,.00008], 'black', linewidth=3)

plt.title('Delta', fontsize=46)
plt.ylabel('mV', fontsize=36)
plt.xlabel('Wavelength in angstroms', fontsize=36)
matplotlib.rc('xtick', labelsize=20)
matplotlib.rc('ytick', labelsize=20)
plt.savefig('delta.png')



print 'D delta = ', mu_sig_opt[1]
print 'H delta = ', mu_sig_opt2[1]
```
